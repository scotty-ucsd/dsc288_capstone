# LEO Index Validation

P0-C1 validates whether the Swarm-derived LEO index explains enough variance in global SuperMAG dB/dt to keep DMSP deferred.

Scientific contract:

- Uses existing feature matrices with per-station `dbdt_horizontal_magnitude_{station}` targets.
- Uses existing LEO index feature columns; no DMSP, MMS, or new source physics is introduced.
- Positive lag means the LEO index leads the SuperMAG target.
- DMSP remains deferred if the best lagged LEO signal explains at least 10% of global SuperMAG dB/dt variance.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from swmi.evaluation.leo_index_validation import run_leo_index_validation

FEATURES_DIR = PROJECT_ROOT / "data" / "processed" / "features"
OUTPUT_DIR = PROJECT_ROOT / "results" / "validation" / "leo_index"

FEATURES_DIR, OUTPUT_DIR
```

## Run Validation

This cell scans lagged correlations between LEO index features and the global mean SuperMAG target, then writes:

- `leo_lag_correlations.csv`
- `leo_station_correlations.csv`
- `leo_lag_scan.png`
- `leo_station_spatial_correlations.png`
- `leo_index_validation_report.md`
- `leo_index_validation_summary.json`

```python
result = run_leo_index_validation(
    features_dir=FEATURES_DIR,
    output_dir=OUTPUT_DIR,
    lags_min=range(-180, 181, 10),
)
result
```
